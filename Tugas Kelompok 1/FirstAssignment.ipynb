#### C. Metodologi yang Digunakan

Metodologi penelitian dalam paper ini adalah studi literatur terstruktur dengan pendekatan konseptual dan komparatif. Penulis menelaah berbagai penelitian terdahulu mengenai penerapan quantum computing pada persoalan utama di bidang keuangan, khususnya optimisasi portofolio, machine learning untuk prediksi pasar, serta simulasi Monte Carlo untuk pricing derivatif dan analisis risiko. Tujuan review ini adalah menilai potensi, efisiensi, keterbatasan, serta peluang integrasi algoritma kuantum ke dalam sistem keuangan klasik.

Berikut langkah-langkah yang dilakukan penulis dalam penelitian:

1. Mengidentifikasi masalah inti dalam quantitative finance  
   Penulis terlebih dahulu memetakan persoalan utama di bidang keuangan, seperti ketidakpastian perilaku aset, pengukuran risiko, optimisasi portofolio, pricing derivatif, serta estimasi return dan risk. Dari sini, paper mengelompokkan tiga pendekatan utama yang umum digunakan, yaitu optimization models, machine learning methods, dan Monte Carlo-based methods.

2. Meninjau landasan teori quantum computing yang relevan  
   Paper kemudian membahas konsep dasar quantum computing, mulai dari prinsip kerja qubit, superposition, entanglement, hingga tahapan komputasi kuantum. Selain itu, dibahas pula algoritma-algoritma penting untuk keuangan, seperti Grover’s algorithm, QAOA, HHL, Quantum PCA, quantum SVM, dan terutama Quantum Amplitude Estimation (QAE) untuk mempercepat simulasi Monte Carlo.

3. Melakukan analisis tematik berdasarkan domain aplikasi  
   Setelah teori dasar dijelaskan, penulis menyusun review secara tematik ke dalam tiga kelompok utama, yaitu Quantum Optimization, Quantum Machine Learning, serta Quantum Amplitude Estimation dan Quantum Monte Carlo. Masing-masing kelompok digunakan untuk membahas penerapan quantum computing pada kasus seperti optimal trading trajectory, arbitrage, credit scoring, classification, regression, PCA, neural networks, derivative pricing, dan risk analysis.

4. Membandingkan penelitian terdahulu untuk menemukan research gap  
   Penulis tidak hanya merangkum literatur, tetapi juga membandingkan fokus, temuan, dan keterbatasan dari studi-studi sebelumnya. Dari analisis ini, ditemukan beberapa celah penelitian, seperti masih terbatasnya implementasi dalam skala nyata, perlunya pendekatan hybrid quantum-classical, dan belum luasnya penerapan simulasi stokastik serta machine learning kuantum pada sistem keuangan yang dinamis.

5. Mensintesis implikasi praktis dari berbagai studi kasus  
   Pada tahap akhir, penulis menyatukan hasil-hasil kajian untuk menunjukkan manfaat praktis quantum computing di sektor keuangan. Sintesis ini menegaskan bahwa metode seperti Quantum Monte Carlo dan QAE berpotensi meningkatkan efisiensi dalam pricing derivatif, pengukuran risiko seperti VaR dan CVaR, serta optimisasi portofolio, meskipun masih menghadapi kendala dalam aspek skalabilitas, kualitas data, integrasi sistem, dan regulasi.
