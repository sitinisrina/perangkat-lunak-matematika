##### 6. Sequence Argument

```python
def doubler(*sequence):
    return [x * 2 for x in sequence]
print(doubler(1, 2, 3, 4, 5))
```

```python
doubler(5, 2, "Wow!")
```

##### 7. Keyword Argument

```python
def arrowify(**args):
    for key, value in args.items():
        print(key + " -> " + value)


arrowify(neutron="n", proton="p", electron="e")
```

```python
def somefunc(a, b, *args, **kwargs):
    print("A:", a)
    print("B:", b)
    print("args:", args)
    print("keyword args", kwargs)
somefunc(1, 2, 3, 4, 5, fish="James")
```
