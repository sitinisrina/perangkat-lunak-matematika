(Hexagon Area Program) Write a program that prompts the user to enter the side of a hexagon and displays its area.  
The formula for computing the area of a hexagon is:

$$
Area = \frac{3\sqrt{3}}{2} s^2

#### 2.1.7

(Science: Wind-Chill Temperature)How cold is it outside? The temperature alone is not enough to provide the answer. Other factors including wind speed, relative humidity, and sunshine play important roles in determining coldness outside.

In 2001, the **National Weather Service (NWS)** implemented the new wind-chill temperature to measure the coldness using temperature and wind speed.

The formula is:

$$
t_{wc} = 35.74 + 0.6215t_a - 35.75v^{0.16} + 0.4275t_av^{0.16}
$$

Where:

- $t_a$ = outside temperature in **degrees Fahrenheit**
- $v$ = wind speed in **miles per hour**
- $t_{wc}$ = **wind-chill temperature**

The formula **cannot be used** when:

- Wind speed is **below 2 mph**
- Temperature is **below −58°F or above 41°F**

### Task

Write a program that:

1. Prompts the user to enter a **temperature between −58°F and 41°F**.
2. Prompts the user to enter a **wind speed greater than or equal to 2 mph**.
3. Computes and displays the **wind-chill temperature**.

Use `math.pow(a, b)` to compute $v^{0.16}$.

```python
# meminta input suhu dari user
t = float(input("Enter the temperature in Fahrenheit between -58°F and 41°F: "))

# meminta input kecepatan angin dari user
v = float(input("Enter the wind speed (>= 2) in miles per hour: "))

# menghitung wind chill temperature
twc = 35.74 + 0.6215 * t - 35.75 * math.pow(v, 0.16) + 0.4275 * t * math.pow(v, 0.16)

# menampilkan hasil
print("The wind chill index is", twc)
```

(Print a Table) Write a program that displays the following table.  
Cast floating-point numbers into integers.

| a | b | pow(a, b) |
|---|---|-----------|
| 1 | 2 | 1 |
| 2 | 3 | 8 |
| 3 | 4 | 81 |
| 4 | 5 | 1024 |
| 5 | 6 | 15625 |

The program should compute the values using the `pow(a, b)` function and display them in table format.


```python
print("a   b   pow(a, b)")
print("-------------------")

for a in range(1, 6):
    b = a + 1
    result = int(math.pow(a, b))
    print(a, " ", b, " ", result)
```

#### 2.1.9

(Geometry: Area of a Triangle) Write a program that prompts the user to enter three points  
$(x_1, y_1)$, $(x_2, y_2)$, and $(x_3, y_3)$ of a triangle and then displays its area.

The formula for computing the area of a triangle is:

$$
s = \frac{side1 + side2 + side3}{2}
$$

$$
area = \sqrt{s(s - side1)(s - side2)(s - side3)}
$$

Where:

- $side1$ = distance between $(x_1, y_1)$ and $(x_2, y_2)$  
- $side2$ = distance between $(x_2, y_2)$ and $(x_3, y_3)$  
- $side3$ = distance between $(x_1, y_1)$ and $(x_3, y_3)$

```python
# meminta input koordinat 1.5 -3.4 4.6 5 9.5 -3.4
x1, y1, x2, y2, x3, y3 = map(float, input(
"Enter the coordinates of three points separated by spaces like x1 y1 x2 y2 x3 y3: "
).split())

# menghitung panjang sisi menggunakan rumus jarak
side1 = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
side2 = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)
side3 = math.sqrt((x3 - x1)**2 + (y3 - y1)**2)

# menghitung s (setengah keliling)
s = (side1 + side2 + side3) / 2

# menghitung luas segitiga (rumus Heron)
area = math.sqrt(s * (s - side1) * (s - side2) * (s - side3))

# menampilkan hasil
print("The area of the triangle is", round(area, 1))
```
