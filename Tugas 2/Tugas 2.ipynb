#### Jawaban:


##### Fungsi Akar Persamaan Kuadrat

Persamaan kuadrat memiliki bentuk umum:

$$
ax^2 + bx + c = 0
$$

Untuk mencari akar-akar dari persamaan kuadrat digunakan rumus diskriminan:

$$
D = b^2 - 4ac
$$

Kemudian akar-akar persamaan dihitung dengan rumus:

$$
x = \frac{-b \pm \sqrt{D}}{2a}
$$

Keterangan kondisi diskriminan:

- Jika $D > 0$ → dua akar **real berbeda**
- Jika $D = 0$ → dua akar **real sama**
- Jika $D < 0$ → akar **kompleks**

Pada notebook ini akan dibuat fungsi Python yang menerima input **a, b, c (float)** dan mengembalikan akar persamaan kuadrat.

```python
## Function untuk menghitung akar kuadrat

def akar_kuadrat(a: float, b: float, c: float):
    
    # menghitung diskriminan
    D = b**2 - 4*a*c
    
    # menghitung akar
    x1 = (-b + cmath.sqrt(D)) / (2*a)
    x2 = (-b - cmath.sqrt(D)) / (2*a)
    
    return x1, x2
```

##### Contoh Penggunaan Fungsi

Berikut contoh kasus untuk menghitung akar kuadrat:

1. **Kasus akar real**

```python
# contoh akar real
a = 1.0
b = -3.0
c = 2.0

akar1, akar2 = akar_kuadrat(a, b, c)

print("Akar real:")
print("x1 =", akar1, "dan x2 =", akar2)
print("Dengan kata lain akarnya adalah: ", akar1.real, "dan", akar2.real)
```

2. **Kasus akar kompleks**

```python
# contoh akar kompleks
a = 1.0
b = 2.0
c = 5.0

akar1, akar2 = akar_kuadrat(a, b, c)

print("Akar kompleks:")
print("x1 =", akar1, "dan x2 =", akar2)
```

#### Menentukan Nilai Minimum Fungsi dengan Metode Biseksi

Metode Biseksi merupakan metode numerik yang digunakan untuk mencari akar suatu fungsi
dalam interval tertentu dengan cara membagi interval menjadi dua bagian secara berulang.

Jika suatu fungsi $f(x)$ memiliki nilai berbeda tanda pada dua titik $a$ dan $b$,
maka terdapat akar di antara interval tersebut.

Pada soal ini metode biseksi digunakan untuk membantu menentukan **nilai minimum**
dari suatu fungsi dalam interval tertentu.

Langkah metode biseksi:

1. Tentukan interval awal $[a,b]$
2. Hitung titik tengah

$$
c = \frac{a + b}{2}
$$

3. Evaluasi fungsi pada titik tersebut
4. Perkecil interval hingga mencapai toleransi yang diinginkan

Program berikut akan membuat fungsi Python untuk melakukan proses tersebut.

```python
## Fucntion untuk metode biseksi
def metode_biseksi(f, a, b, tol=1e-6, max_iter=100):
    
    if f(a) * f(b) >= 0:
        print("Interval tidak memenuhi syarat metode biseksi")
        return None
    
    for i in range(max_iter):
        
        c = (a + b) / 2
        
        if abs(f(c)) < tol:
            return c
        
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
            
    return (a + b) / 2
```

#### Contoh Kasus

Akan diberikan dua contoh fungsi $f(x)$ untuk mencari nilai minimum menggunakan metode biseksi.

```python
# contoh pertama
def f1(x):
    return x**2 - 4

akar1 = metode_biseksi(f1, 0, 3)

print("Kasus 1")
print("Nilai minimum mendekati:", akar1)
```

```python
# contoh kedua
def f2(x):
    return x**3 - x - 2

akar2 = metode_biseksi(f2, 1, 2)

print("Kasus 2")
print("Nilai minimum mendekati:", akar2)
```

#### Menentukan Nilai Minimum Fungsi dengan Metode Newton-Raphson

Metode Newton-Raphson merupakan metode numerik yang digunakan untuk mencari
akar suatu fungsi dengan pendekatan iteratif.

Rumus dasar metode Newton-Raphson adalah:

$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$

Metode ini menggunakan turunan pertama dari fungsi untuk mempercepat
proses konvergensi menuju solusi.

Pada soal ini metode Newton-Raphson digunakan untuk mencari nilai minimum
suatu fungsi dengan mencari akar dari turunan pertama fungsi tersebut.

Program akan membuat fungsi Python yang melakukan iterasi Newton-Raphson
hingga mencapai toleransi tertentu atau hingga jumlah iterasi maksimum
(100 iterasi).

```python
## Function untuk metode Newton-Raphson
def newton_raphson(f, df, x0, tol=1e-6, max_iter=100):

    x = x0

    for i in range(max_iter):

        if df(x) == 0:
            print("Turunan bernilai nol, metode gagal.")
            return None

        x_new = x - f(x)/df(x)

        if abs(x_new - x) < tol:
            return x_new

        x = x_new

    print("Metode tidak konvergen.")
    return None
```

#### Jawaban:

### Estimasi Parameter Distribusi Negatif Binomial dengan Metode Newton-Raphson

Diasumsikan data mengikuti distribusi **Negatif Binomial** dengan fungsi massa probabilitas:

$$
p(X=x \mid p,r)=\binom{x+r-1}{r-1} p^x (1-p)^r,
\qquad 0<p<1,\; r=1,2,3,\dots
$$

Untuk sampel acak $x_1,x_2,\dots,x_n$, fungsi likelihood adalah:

$$
L(p,r)=\prod_{i=1}^{n} \binom{x_i+r-1}{r-1} p^{x_i}(1-p)^r
$$

Karena lebih mudah dihitung, digunakan **log-likelihood**:

$$
\ell(p,r)=\sum_{i=1}^{n} \left[
\ln \Gamma(x_i+r)-\ln \Gamma(r)-\ln(x_i!)
+ x_i \ln p + r \ln(1-p)
\right]
$$

dengan $\Gamma(\cdot)$ adalah fungsi gamma.

---

### Gradien

Turunan pertama terhadap $p$:

$$
\frac{\partial \ell}{\partial p}
=
\sum_{i=1}^{n} \frac{x_i}{p}
-
\frac{nr}{1-p}
$$

Turunan pertama terhadap $r$:

$$
\frac{\partial \ell}{\partial r}
=
\sum_{i=1}^{n}
\left[
\psi(x_i+r)-\psi(r)
\right]
+
n \ln(1-p)
$$

dengan $\psi(\cdot)$ adalah fungsi **digamma**.

---

### Matriks Hessian

Turunan kedua terhadap $p$:

$$
\frac{\partial^2 \ell}{\partial p^2}
=
-
\sum_{i=1}^{n} \frac{x_i}{p^2}
-
\frac{nr}{(1-p)^2}
$$

Turunan silang:

$$
\frac{\partial^2 \ell}{\partial p \partial r}
=
\frac{\partial^2 \ell}{\partial r \partial p}
=
-\frac{n}{1-p}
$$

Turunan kedua terhadap $r$:

$$
\frac{\partial^2 \ell}{\partial r^2}
=
\sum_{i=1}^{n}
\left[
\psi_1(x_i+r)-\psi_1(r)
\right]
$$

dengan $\psi_1(\cdot)$ adalah fungsi **trigamma**.

Sehingga matriks Hessian dapat ditulis sebagai:

$$
H(p,r)=
\begin{bmatrix}
\frac{\partial^2 \ell}{\partial p^2} & \frac{\partial^2 \ell}{\partial p \partial r} \\
\frac{\partial^2 \ell}{\partial r \partial p} & \frac{\partial^2 \ell}{\partial r^2}
\end{bmatrix}
$$

---

### Metode Newton-Raphson

Parameter diperbarui dengan rumus:

$$
\theta^{(k+1)}=\theta^{(k)}-H(\theta^{(k)})^{-1} g(\theta^{(k)})
$$

dengan:

$$
\theta=
\begin{bmatrix}
p \\
r
\end{bmatrix}
,\qquad
g(\theta)=
\begin{bmatrix}
\frac{\partial \ell}{\partial p} \\
\frac{\partial \ell}{\partial r}
\end{bmatrix}
$$

Pada program ini, input yang digunakan adalah:

- data,
- nilai awal $p$,
- nilai awal $r$,
- maksimum iterasi sebanyak $100$.

```python
data = [
    [8, 6, 4, 9],
    [6, 14, 14, 7],
    [12, 18, 7, 6],
    [6, 6, 9, 20],
    [10, 15, 7, 15],
    [21, 16, 8, 4],
    [9, 9, 15, 6],
    [9, 5, 10, 9],
    [12, 6, 2, 12],
    [8, 10, 13, 3],
    [4, 9, 4, 6],
    [13, 5, 7, 10],
    [15, 5, 11, 10],
    [15, 9, 9, 11],
    [12, 11, 9, 15],
    [6, 6, 15, 5],
    [2, 12, 6, 10],
    [5, 12, 8, 3],
    [6, 15, 16, 8],
    [5, 7, 2, 6],
    [7, 10, 15, 8],
    [13, 14, 3, 13],
    [10, 15, 5, 13],
    [13, 11, 5, 16],
    [14, 5, 17, 21]
]

# ubah menjadi vektor 1 dimensi
x = [nilai for baris in data for nilai in baris]
print("Jumlah data =", len(x))
print("5 data pertama =", x[:5])
```

```python
#Fungsi Log-Likelihood, Gradien, dan Hessian untuk distribusi Negative Binomial

def log_likelihood_nb(x, p, r):
    x = np.array(x, dtype=float)

    if p <= 0 or p >= 1 or r <= 0:
        return -np.inf

    ll = np.sum(
        gammaln(x + r)
        - gammaln(r)
        - gammaln(x + 1)
        + x * np.log(p)
        + r * np.log(1 - p)
    )
    return ll

def gradient_nb(x, p, r):
    x = np.array(x, dtype=float)
    n = len(x)

    dlp = np.sum(x) / p - (n * r) / (1 - p)
    dlr = np.sum(digamma(x + r) - digamma(r)) + n * np.log(1 - p)

    return np.array([dlp, dlr], dtype=float)

def hessian_nb(x, p, r):
    x = np.array(x, dtype=float)
    n = len(x)

    d2pp = -np.sum(x) / (p**2) - (n * r) / ((1 - p)**2)
    d2pr = -n / (1 - p)
    d2rr = np.sum(polygamma(1, x + r) - polygamma(1, r))

    H = np.array([
        [d2pp, d2pr],
        [d2pr, d2rr]
    ], dtype=float)

    return H
```

```python
#Fungsi untuk metode Newton-Raphson untuk estimasi MLE

def newton_raphson_nb(x, p0, r0, tol=1e-8, max_iter=100):
    p = float(p0)
    r = float(r0)

    for i in range(max_iter):
        grad = gradient_nb(x, p, r)
        hess = hessian_nb(x, p, r)

        try:
            delta = np.linalg.solve(hess, grad)
        except np.linalg.LinAlgError:
            print("Matriks Hessian singular, metode berhenti.")
            return None

        p_new = p - delta[0]
        r_new = r - delta[1]

        # menjaga agar parameter tetap valid
        langkah = 1.0
        ll_lama = log_likelihood_nb(x, p, r)

        while not (0 < p_new < 1 and r_new > 0):
            langkah /= 2
            p_new = p - langkah * delta[0]
            r_new = r - langkah * delta[1]
            if langkah < 1e-10:
                print("Langkah terlalu kecil, metode berhenti.")
                return None

        ll_baru = log_likelihood_nb(x, p_new, r_new)

        while ll_baru < ll_lama:
            langkah /= 2
            p_new = p - langkah * delta[0]
            r_new = r - langkah * delta[1]

            if not (0 < p_new < 1 and r_new > 0) or langkah < 1e-10:
                print("Gagal menemukan langkah yang valid.")
                return None

            ll_baru = log_likelihood_nb(x, p_new, r_new)

        if np.linalg.norm([p_new - p, r_new - r], ord=2) < tol:
            return {
                "p_hat": p_new,
                "r_hat": r_new,
                "iterasi": i + 1,
                "log_likelihood": ll_baru
            }

        p, r = p_new, r_new

    return {
        "p_hat": p,
        "r_hat": r,
        "iterasi": max_iter,
        "log_likelihood": log_likelihood_nb(x, p, r)
    }
```

Pada contoh ini digunakan nilai awal:

- $p_0 = 0.5$
- $r_0 = 9$
- maksimum iterasi $= 100$

```python
#pemanggilan fungsi

hasil = newton_raphson_nb(x, p0=0.5, r0=9, max_iter=100)

print("Hasil Estimasi MLE")
print("p_hat          =", round(hasil["p_hat"], 6))
print("r_hat          =", round(hasil["r_hat"], 6))
print("Jumlah iterasi =", hasil["iterasi"])
print("Log-likelihood =", round(hasil["log_likelihood"], 6))
```

#### Jawaban:

#### Membangkitkan Bilangan Gompertz dengan Metode Inverse Transform

Pada soal ini akan dibuat fungsi untuk membangkitkan bilangan acak
berdistribusi Gompertz dari bilangan acak Uniform $(0,1)$ menggunakan
metode **Inverse Transform**.

CDF yang digunakan mengikuti soal, yaitu:

$$
F(x)=1-\exp\left(\eta(e^{bx}-1)\right)
$$

dengan parameter default:

$$
\eta = 2
$$

$$
b = 1.5
$$

---

#### Menurunkan Fungsi Invers

Misalkan:

$$
U = F(X)
$$

maka:

$$
U = 1-\exp\left(\eta(e^{bX}-1)\right)
$$

Sehingga:

$$
1-U = \exp\left(\eta(e^{bX}-1)\right)
$$

Ambil logaritma natural:

$$
\ln(1-U)=\eta(e^{bX}-1)
$$

$$
e^{bX}=1+\frac{\ln(1-U)}{\eta}
$$

Ambil logaritma kembali:

$$
bX=\ln\left(1+\frac{\ln(1-U)}{\eta}\right)
$$

Jadi fungsi inversnya adalah:

$$
X=\frac{1}{b}\ln\left(1+\frac{\ln(1-U)}{\eta}\right)
$$

---

#### Syarat Nilai Uniform

Agar fungsi logaritma terdefinisi, harus berlaku:

$$
1+\frac{\ln(1-U)}{\eta}>0
$$

yang ekuivalen dengan:

$$
U<1-e^{-\eta}
$$

Untuk $\eta=2$ diperoleh:

$$
U<1-e^{-2}\approx 0.8646647
$$

Jadi pada program ini, hanya nilai Uniform yang memenuhi syarat tersebut
yang akan digunakan dalam transformasi.

```python
## Fungsi Pembangkit
def gompertz_inverse_transform_soal(n, eta=2, b=1.5, seed=None):
    if n <= 0:
        raise ValueError("n harus lebih besar dari 0")
    if eta <= 0 or b <= 0:
        raise ValueError("eta dan b harus lebih besar dari 0")

    if seed is not None:
        np.random.seed(seed)

    batas_u = 1 - np.exp(-eta)

    u_valid = []

    # membangkitkan U ~ Uniform(0,1) sampai terkumpul n nilai yang valid
    while len(u_valid) < n:
        u = np.random.uniform(0, 1)
        if u < batas_u:
            u_valid.append(u)

    u_valid = np.array(u_valid)

    # transformasi invers sesuai soal
    x = (1 / b) * np.log(1 + (np.log(1 - u_valid) / eta))

    return u_valid, x
```

#### Contoh Penggunaan

Pada contoh ini digunakan:

- $n = 10$
- $\eta = 2$
- $b = 1.5$

```python
#Pemanggilan fungsi

u, x = gompertz_inverse_transform_soal(n=10, eta=2, b=1.5, seed=42)

print("Bilangan Uniform(0,1) yang valid:")
print(u)

print("\nBilangan hasil transformasi Gompertz:")
print(x)
```
