### 1. Conditionality

```python
x = -2.345
if x < 0:
    print(x, "is negative")
elif x == 0:
    print(x,"is zero")
else:
    print(x, "is positive")
```

```python
choice = "High" #python is case sensitive

if choice == "high":
    print(1)
elif choice == "medium":
    print(2)
else:
    print(3)
```

## 4. Iteration

```python
mylist = [3, 7, 15, 2]
for whatever in mylist:
    print(whatever**2)
```

```python
vowels = "aeiou"
sarcasm = []

for letter in "Okay":
    if letter.lower() in vowels:
        repetition = 3
    else:
        repetition = 1

    sarcasm.append(letter * repetition)

"".join(sarcasm)
```

```python
current_year = 2022
founded = {"Barack Obama": 1961, "UCL": 1826, "The Alan Turing Institute": 2015}

for thing in founded:
    print(f"In {current_year} {thing} is {current_year - founded[thing]} years old.")
```

## 5. Defining Method

```python
class Room:
    def overfull(self):
        return len(self.occupants) > self.capacity
```

```python
myroom = Room()
myroom.capacity = 3
myroom.occupants = ["James", "Sue"]
```

```python
myroom.overfull()
```
