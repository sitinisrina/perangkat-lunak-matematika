```python
import random

# Generate lottery number (3 digit)
lottery = random.randint(100, 999)

# Input user number
user_input = int(input("Masukkan angka lotre 3 digit: "))

# Convert to string to compare digits
lottery_str = str(lottery)
user_str = str(user_input)

print("Angka lotre adalah:", lottery)

# Rule checking
if user_str == lottery_str:
    print("Selamat! Anda menang $10,000")
    
elif sorted(user_str) == sorted(lottery_str):
    print("Selamat! Semua digit cocok. Anda menang $3,000")
    
elif any(digit in lottery_str for digit in user_str):
    print("Ada satu digit yang cocok. Anda menang $1,000")
    
else:
    print("Maaf, Anda tidak menang.")
```
