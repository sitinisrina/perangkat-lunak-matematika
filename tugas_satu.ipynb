## Nomor 1
Soal: Jelaskan jenis-jenis error di Python, dan berikan contohnya disertai dengan cara menangani error tersebut.

Jawaban:

#### 1. Syntax Error

Syntax Error adalah kesalahan dalam penulisan kode yang tidak sesuai dengan aturan (syntax) Python.
Error ini muncul sebelum program dijalankan.

```python
# Contoh Syntax Error
if 5 > 3
    print("Benar")
```

Output menunjukkan peringatan syntax error, karena kurang tanda titik dua setelah kondisi if. 

Contoh lain syntax indentasi, untuk conditional statement, python cukup strict soal indentasi.


#### 2. Runtime Error (Exception)

Exception adalah error yang terjadi saat program sedang dijalankan (runtime).
Biasanya disebabkan oleh kesalahan logika atau input yang tidak sesuai.

Berikut terdapat beberapa contoh exception yang umum terjadi.


##### a. ZeroDivisionError

```python
# Contoh ZeroDivisionError
# angka = 10
# hasil = angka / 0
# print(hasil)

try:
    angka = 10
    hasil = angka / 0
    print(hasil)
except ZeroDivisionError:
    print("Terjadi pembagian dengan nol!")
```

Sehingga error tersebut ditangkap oleh Exception ZeroDivisorError.

##### b. IndexError

```python
# Contoh IndexError
# data = [1, 2, 3]
# print(data[5])

try:
    data = [1, 2, 3]
    print(data[5])
except IndexError:
    print("Index tidak ditemukan dalam list!")
```

Error muncul karena user ingin print index ke-5, sementara yang diinisialisasikan hanya sampai index ke-2 (dimulai dari 0), sehingga error ditangkap oleh Exception IndexError.

#### 3. Logical Error 

Logical Error adalah kesalahan logika dalam program.
Program tetap berjalan tanpa error, tetapi hasilnya tidak sesuai dengan yang diharapkan.


```python
# Contoh Logical Error
panjang = 5
lebar = 4

luas = panjang + lebar
print("Luas:", luas)
```

Rumus luas harunya panjang x lebar, sehingga terdapat kesalahan logika disini karena salah tanda operator.

#### 4. Struktur Dasar Penanganan Error

Python menyediakan beberapa blok untuk menangani error:

- try berisi kode yang akan diuji,

- except untuk menangkap error pada kode yang diuji,

- else melanjutkan jalan program jika tidak ada error,

- finally selalu dijalankan untuk memastikan program selesai.


```python
# contoh lengkap
try:
    angka = 7
    hasil = 10 / angka
except ZeroDivisionError:
    print("Tidak boleh membagi dengan nol!")
except ValueError:
    print("Input harus berupa angka!")
else:
    print("Hasil:", hasil)
finally:
    print("Program selesai.")
```

## Nomor 4
Soal: Buatlah program yang dapat menampilkan output berupa akar-akar dari $z^6=8$. (Petunjuk: Ingat kembali materi mencari akar dari bilangan kompleks.)

Jawaban:

Diketahui:

$$
z^6 = 8
$$

Bilangan 8 dapat ditulis dalam bentuk polar sebagai:

$$
8 = 8 \, \text{cis}(0)
$$

Rumus akar ke-n dari bilangan kompleks:

$$
z_k = r^{1/n} \, \text{cis}\left(\frac{\theta + 2k\pi}{n}\right)
$$

Dengan:
- r = 8
- n = 6
- k = 0, 1, 2, 3, 4, 5

Sehingga:

$$
z_k = 8^{1/6} \, \text{cis}\left(\frac{2k\pi}{6}\right)
$$

```python
# nilai yang diketahui
r = 8
n = 6

# menghitung akar ke-6 dari 8
akar_pangkat = r ** (1/n)

print("Akar-akar dari z^6 = 8 adalah:\n")

for k in range(n):
    sudut = (2 * math.pi * k) / n
    z = akar_pangkat * cmath.exp(1j * sudut)
    print(f"z{k} =", z)
```

Program dimulai dengan mengimpor library `cmath` untuk operasi bilangan kompleks dan `math` untuk nilai π. 
Nilai r = 8 dan n = 6 disimpan sebagai dasar perhitungan akar.

Kemudian dihitung nilai 8^(1/6), yaitu akar keenam dari 8. 
Selanjutnya program menggunakan rumus akar bilangan kompleks:

$$
z_k = r^{\frac{1}{n}} \cdot e^{i\left(\frac{2k\pi}{n}\right)}
$$

Program melakukan perulangan sebanyak 6 kali (k = 0 sampai 5) untuk mendapatkan seluruh akar kompleksnya. 
Setiap nilai sudut dihitung menggunakan rumus $(2πk)/6$, lalu dikonversi ke bentuk kompleks menggunakan fungsi `cmath.exp()`.

Hasil akhirnya menampilkan 6 akar kompleks yang memenuhi persamaan z⁶ = 8.

## Nomor 5
Soal: Buatlah program yang dapat menampilkan output berupa $10$ bilangan pertama dari barisan Fibonacci 
$$
f_n=f_{n-1}+f_{n-2}.
$$
(Petunjuk: $f_0=f_1=1, f_2=2$ dengan $f_n$ adalah bilangan Fibonacci ke-$n$.)

Jawaban:

Barisan Fibonacci didefinisikan dengan rumus:

$$
f_n = f_{n-1} + f_{n-2}
$$

Dengan kondisi awal:
- $f_0 = 1$
- $f_1 = 1$

Diminta menampilkan 10 bilangan pertama dari barisan tersebut.

```python
# jumlah bilangan Fibonacci yang ingin ditampilkan
n = 10

# nilai awal
f0 = 1
f1 = 1

print("10 bilangan pertama Fibonacci:")

print(f0)
print(f1)

for i in range(2, n):
    fn = f0 + f1
    print(fn)
    f0 = f1
    f1 = fn
```

Program menentukan jumlah bilangan Fibonacci yang ingin ditampilkan sebanyak 10. 
Kemudian ditetapkan dua nilai awal, yaitu $f_0 = 1$ dan $f_1 = 1$.

Dua nilai awal tersebut langsung ditampilkan terlebih dahulu. 
Selanjutnya program melakukan perulangan mulai dari indeks ke-2 sampai ke-9. 

Pada setiap perulangan, nilai Fibonacci berikutnya dihitung menggunakan rumus:

$$
f_n = f_{n-1} + f_{n-2}
$$

Nilai baru disimpan dalam variabel $`f_n`$, kemudian ditampilkan. 
Setelah itu, nilai sebelumnya diperbarui agar proses dapat berlanjut hingga 10 bilangan pertama berhasil ditampilkan.

## Nomor 6
Soal: Buatlah program yang dapat menampilkan output berupa $5$ jumlahan parsial pertama dari deret 
$$
\sum_{n=0}^\infty \frac{1}{n^2}.
$$

Jawaban:

Diketahui deret:

$$
\sum_{n=0}^{\infty} \frac{1}{n^2}
$$

Jumlah parsial adalah penjumlahan dari beberapa suku pertama deret tersebut.

Diminta menampilkan 5 jumlah parsial pertama.

```python
# jumlah parsial yang ingin ditampilkan
jumlah_parsial = 5

total = 0

print("5 jumlah parsial pertama:")

for n in range(0, jumlah_parsial):
    total += 1 / (n**2)   # saat n = 0 akan terjadi error
    print(f"S{n} =", total)
```

Program melakukan perulangan mulai dari n = 0 sesuai dengan batas bawah deret.

Pada iterasi pertama (n = 0), program mencoba menghitung:

$$
\frac{1}{0^2}
=
\frac{1}{0}
$$

Operasi tersebut tidak terdefinisi dalam matematika karena pembagian dengan nol tidak diperbolehkan.

Akibatnya, Python melempar exception berupa: ZeroDivisionError.

Exception ini muncul karena program mencoba melakukan operasi pembagian dengan nol sebelum jumlah parsial dapat dihitung.

## Nomor 7
Soal: Buatlah program yang dapat menampilkan output berupa bentuk bilangan desimal dari bilangan biner
$$
1111111111111000000000000.
$$
Apa fakta menarik yang Anda ketahui tentang bilangan ini?

Jawaban:

Diketahui bilangan biner:

$$
1111111111111000000000000
$$

Program diminta untuk mengubah bilangan biner tersebut ke dalam bentuk desimal.

```python
# bilangan biner dalam bentuk string
biner = "1111111111111000000000000"

# konversi ke desimal
desimal = int(biner, 2)

print("Bilangan biner :", biner)
print("Dalam desimal  :", desimal)
```

##### Penjelasan Program

Program menyimpan bilangan biner dalam bentuk string agar dapat diproses oleh fungsi konversi. 
Kemudian fungsi `int(biner, 2)` digunakan untuk mengubah bilangan berbasis 2 (biner) menjadi bilangan desimal. 
Angka 2 menunjukkan bahwa sistem bilangan yang digunakan adalah basis 2.

Hasil konversi kemudian ditampilkan menggunakan perintah `print()`.

##### Fakta Menarik

Bilangan biner tersebut terdiri dari beberapa digit 1 yang diikuti oleh nol. 
Bentuk seperti ini dalam sistem biner merepresentasikan bilangan yang sangat berkaitan dengan pangkat dua.

Secara umum, pola:
111...11000...000

dapat ditulis sebagai:

$$
(2^n - 1) × 2^k
$$

Artinya bilangan ini merupakan hasil dari selisih satu dari suatu pangkat dua, 
kemudian dikalikan lagi dengan pangkat dua lainnya.

Bilangan dengan pola seperti ini sering muncul dalam representasi memori komputer dan sistem digital.

Soal: Buatlah program yang dapat menampilkan output berupa titik berat dari bidang datar segitiga dengan titik sudut di $(0,0), (4,6), (4,-6)$. (Petunjuk: Ingat kembali materi titik berat di Kalkulus 2)

Jawaban:

Diketahui titik sudut segitiga:
- A $(0, 0)$
- B $(4, 6)$
- C $(4, -6)$

Rumus titik berat (centroid) segitiga adalah:

$$
\bar{x} = \frac{x_1 + x_2 + x_3}{3}
$$

$$
\bar{y} = \frac{y_1 + y_2 + y_3}{3}
$$

```python
# koordinat titik-titik segitiga
x1, y1 = 0, 0
x2, y2 = 4, 6
x3, y3 = 4, -6

# menghitung titik berat
x_bar = (x1 + x2 + x3) / 3
y_bar = (y1 + y2 + y3) / 3

print("Titik berat segitiga adalah:")
print("(", x_bar, ",", y_bar, ")")
```

Program terlebih dahulu menyimpan koordinat ketiga titik sudut segitiga ke dalam variabel.

Kemudian titik berat dihitung menggunakan rumus:

$$
\bar{x} = \frac{x_1 + x_2 + x_3}{3}
$$

$$
\bar{y} = \frac{y_1 + y_2 + y_3}{3}
$$

Nilai rata-rata dari ketiga koordinat x menghasilkan $\bar{x}$, 
dan rata-rata dari ketiga koordinat y menghasilkan $\bar{y}$.

Hasil perhitungan kemudian ditampilkan sebagai koordinat titik berat segitiga.
